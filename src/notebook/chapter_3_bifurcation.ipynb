# Chapter 3 -- Bifurcation

```python
import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = np.arange(0, 200)
n_sample = 5000
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
cmap = plt.get_cmap("spring")
f = lambda a, b, x: x*(1-x*x)-a*(1-np.exp(-b*x))
x = -1.5 + 3 * np.random.random(n_sample)
for a, b in zip([-2, -1, 2.0, 4.0, 3.0], [0.5, 0, 0.5, 0.5, 1.0]):
    ax.scatter(x, f(a, b, x), s=0.2, color=cmap(a*b/2), label=rf"$ab={a*b}$")
# ax.scatter(x, f(-1,0.5, x), s=0.8, color="tab:orange")
ax.axhline(y=0, color="black")
ax.axvline(x=0, color="black")
ax.grid()
ax.legend()
```

```python
f = lambda r, x: 1+r*x+x*x
f_plus = lambda r: (-r+np.sqrt(r**2-4))/2
f_minus = lambda r: (-r-np.sqrt(r**2-4))/2

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
rs = np.linspace(-3, -2, 100)
ax.plot(rs, f_plus(rs), linestyle="-.", color='tab:orange', label="unstable fixed points")
ax.plot(rs, f_minus(rs), linestyle="-", color='tab:orange', label="stable fixed points")

rs = np.linspace(2, 3, 100)
ax.plot(rs, f_plus(rs), linestyle="-.", color='tab:orange')
ax.plot(rs, f_minus(rs), linestyle="-", color='tab:orange')
ax.axhline(y=0, color="black")
ax.axvline(x=0, color="black")
ax.legend()
ax.grid()
```
