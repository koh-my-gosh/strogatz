# Chapter 6 Phase Plane

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.extend(["../library"])

from core import strogatz
```

```python
F = [
    lambda x: x[..., 0] + np.exp(-x[..., 1]),
    lambda x: -x[..., 1]
]
system = strogatz.dynamics2D(F=F, xlim=(-3, 5), ylim=(-2, 3))
system.add_direction_field()
system.add_nullclines()
system.add_fixed_points(fixed_points=[[-1, 0]], fp_types=["saddle"], markersize=18)
system.add_flows(initials=[[1, 3], [-3, -1.5], [-2.5, -2], [-1.5, 0.01], [-1.0, 3.0], [-1.0, -0.2]])
```
