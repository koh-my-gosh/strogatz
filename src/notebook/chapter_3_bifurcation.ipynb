```python
import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda a, b, x: x*(1-x*x)-a*(1-np.exp(-b*x))
```

```python
T = 1000

rec = np.zeros((T, n_sample))
x = -1 + 2 * np.random.random(n_sample)

for t in tqdm.tqdm(range(T)):
    rec[t] = x
    x = f(0.5 , 2.1, x)
```

```python
ts = np.arange(0, 200)
n_sample = 5000
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
cmap = plt.get_cmap("spring")
x = -1.5 + 3 * np.random.random(n_sample)
for a, b in zip([-2, -1, 2.0, 4.0, 3.0], [0.5, 0, 0.5, 0.5, 1.0]):
    ax.scatter(x, f(a, b, x), s=0.2, color=cmap(a*b/2), label=rf"$ab={a*b}$")
# ax.scatter(x, f(-1,0.5, x), s=0.8, color="tab:orange")
ax.axhline(y=0, color="black")
ax.axvline(x=0, color="black")
ax.grid()
ax.legend()
```
