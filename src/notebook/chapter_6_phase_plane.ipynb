```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.extend(["../library"])

from core import strogatz
```

```python
F = [
    lambda x: x[..., 0] + np.exp(-x[..., 1]),
    lambda x: -x[..., 1]
]
system = strogatz.dynamics2D(F=F, xlim=(-3, 5), ylim=(-2, 3), fig_scale=1.5)
system.add_direction_field()
system.add_nullclines()
system.add_fixed_points(fixed_points=[[-1, 0]], fp_types=["saddle"], markersize=18)
system.add_flows(initials=[[1, 3], [-3, -1.5], [-2.5, -2], [-1.5, 0.01], [-1.0, 3.0], [-1.0, -0.2]])
```

```python
F = [
    lambda x: -x[..., 0] + x[..., 0]**3,
    lambda x: -2*x[..., 1]
]
system = strogatz.dynamics2D(F=F, xlim=(-3, 5), ylim=(-2, 3), fig_scale=1.5)
system.add_direction_field()
system.add_nullclines()
system.add_fixed_points(fixed_points=[[-1, 0], [0, 0], [1, 0]], fp_types=["saddle", "stable", "saddle"], markersize=18)
system.add_flows(initials=[[1.1, 3], [1.1, -2], [-1.1, 3], [-1.1, -2], [0.5, -2], [-0.5, -2], [0.5, -2], [-0.5, -2], [0.5, 2], [-0.5, 2]])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
for idx, a in enumerate([-2, 0, 2]):
    F = [
        lambda x: -x[..., 1] + a*x[..., 0]*(x[..., 0]**2+x[..., 1]**2),
        lambda x: x[..., 0]  + a*x[..., 1]*(x[..., 0]**2+x[..., 1]**2),
    ]
    system = strogatz.dynamics2D(F=F, xlim=(-3, 3), ylim=(-3, 3), fig_scale=1.5, ax=ax[idx])
    system.add_direction_field()
    # system.add_nullclines()
    system.add_fixed_points(fixed_points=[[0, 0]], fp_types=["stable"], markersize=18)
    system.add_flows(initials=[[1.1, 3], [1.1, -2], [-1.1, 3], [-1.1, -2], [0.5, -2], [-0.5, -2], [0.5, -2], [-0.5, -2], [0.5, 2], [-0.5, 2]])
```

```python
F = [
    lambda x: x[..., 0]*(3-x[..., 0]-2*x[..., 1]), 
    lambda x: x[..., 1]*(2-x[..., 0]-x[..., 1]), 
]
system = strogatz.dynamics2D(F=F, xlim=(0, 4), ylim=(0, 3), fig_scale=2.0, n_interval=10)
system.add_direction_field()
# system.add_nullclines()
system.add_fixed_points(fixed_points=[[0, 0], [0, 2], [3, 0], [1, 1]], fp_types=["unstable", "stable", "stable", "saddle"], markersize=18)
system.add_flows(initials=[[0.5, 1], [1, 0.5], [1.0, 0.9], [0.8, 1.0], [1.1, 1.0], [2, 1], [1, 1.5]], max_length=10)
# system.add_flows(initials=None, max_length=10)
```

```python
F = [
    lambda x: x[..., 1], 
    lambda x: x[..., 0]-x[..., 0]**3, 
]
system = strogatz.dynamics2D(F=F, xlim=(-2, 2), ylim=(-2, 2), fig_scale=2.0, n_interval=10)
system.add_direction_field()
system.add_nullclines()
system.add_fixed_points(fixed_points=[[-1, 0], [0, 0], [1, 0]], fp_types=["stable", "saddle", "stable"], markersize=18)
system.add_flows(initials=[[0.5, 1], [1, 0.5], [1.0, 0.9], [0.8, 1.0], [-0.3, -0.3]], max_length=10)
# system.add_flows(initials=None, max_length=10)
```

```python
F = [
    lambda x: x[..., 1]-x[..., 1]**3, 
    lambda x: -x[..., 0]-x[..., 1]**2, 
]
system = strogatz.dynamics2D(F=F, xlim=(-2, 1), ylim=(-2, 2), fig_scale=1.5, n_interval=10)
system.add_direction_field()
# system.add_nullclines()
system.add_fixed_points(fixed_points=[[-1, 1], [0, 0], [-1, -1]], fp_types=["saddle", "stable", "saddle"], markersize=18)
# system.add_flows(initials=[[0.5, 1], [-0.8, 1.0], [-0.5, 0.5], [-1.5, 0.0]], max_length=10)
system.add_flows(initials=None, max_length=10)
```
